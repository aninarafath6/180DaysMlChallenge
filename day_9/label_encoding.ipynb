## **Label Encoding**
#### `Data preprocessing`

```python
import numpy as np
import pandas as pd

# label encoding function
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("data/data.csv")
df.head()
```

```python
df["diagnosis"].value_counts()
```

```python
#Load label encoder function
label_encode = LabelEncoder()
labels = label_encode.fit_transform(df["diagnosis"])
```

```python
# appending labels to data frame
df["target"] = labels

df.head()
```

```python
df.target.value_counts()
```

```python
iris_df = pd.read_csv('data/iris_data.csv')
iris_df.head()
```

```python
iris_df.Species.value_counts()
```

```python
iris_df["target"] = label_encode.fit_transform(iris_df["Species"])
iris_df.head()
```
